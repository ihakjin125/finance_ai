```python
import pandas as pd

stock_names = [
    "Samsung",
    "SKHynix",
    "DosanEner",
    "Kodex200"
]
```

```python
for name in stock_names:
    df = pd.read_csv(f"../data/{name}_stock.csv", index_col=0)

    first_price = df["Close"].iloc[0]
    last_price = df["Close"].iloc[-1]
    return_rate = (last_price - first_price) / first_price * 100

    print(f"{name}: {return_rate:.2f}%")
```

```python
results = []
results.append(["Samsung",89.73])
results.append(["SKHynix", 102.35])
results.append(["DosanEner", -17.64])
results.append(["Kodex200", 71.81])
print(results)
```

```python
result_df = pd.DataFrame(
    results,
    columns=["Stock", "Return (%)"]
)

result_df
```

```python
result_df = result_df.sort_values(
    by="Return (%)",
    ascending=False
)

result_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(
    result_df["Stock"],
    result_df["Return (%)"]
)
for stock, ret in zip(result_df["Stock"], result_df["Return (%)"]):

    plt.text(
        stock,
        ret,
        f"{ret:.2f}%",
        ha="center",
        va="bottom"
    )

plt.title("Stock Return Comparison")
plt.xlabel("Stock")
plt.ylabel("Return (%)")

plt.show()
```
