```python
import pandas as pd

stock_names = [
    "Samsung",
    "SKHynix",
    "DosanEner",
    "Kodex200"
]

df = pd.read_csv("../data/Samsung_stock.csv", index_col=0)
daily_return = df["Close"].pct_change()

daily_return.head()
```

```python
average_return = daily_return.mean()
print(f"평균: {average_return * 100:.2f}%")
```

```python
volatility = daily_return.std()
print(f"변동성: {volatility * 100:.2f}%")
```

```python
results = []

for name in stock_names:
    df = pd.read_csv(f"../data/{name}_stock.csv", index_col=0)

    daily_return = df["Close"].pct_change()

    average_return = daily_return.mean() * 100
    volatility = daily_return.std() * 100

    results.append([
        name,
        average_return,
        volatility
    ])

result_df = pd.DataFrame(
    results,
    columns=["Stock", "Average Return (%)", "Volatility (%)"]
)

result_df
```

```python
result_df["Risk Adjusted Return"] = (
    result_df["Average Return (%)"] /
    result_df["Volatility (%)"]
)

result_df
```
