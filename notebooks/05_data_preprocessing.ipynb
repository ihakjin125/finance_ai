```python
import pandas as pd

stock_names = [
    "Samsung",
    "SKHynix",
    "DosanEner",
    "Kodex200"
]

df = pd.read_csv(
    "../data/Samsung_stock.csv",
    index_col = 0
)

close_df = df[["Close"]]

close_df.head()
```

```python
close_df = close_df.rename(
    columns={
        "Close":"Samsung"
    }
)

close_df.head()
```

```python
stock_data = {}

for name in stock_names:
    df = pd.read_csv(
        f"../data/{name}_stock.csv",
        index_col=0
    )

    close_df = df[["Close"]]

    close_df = close_df.rename(
        columns={
            "Close":name
        }
    )

    stock_data[name] = close_df
    print(stock_data.keys())
```

```python
stock_data["Samsung"].head()
```

```python
merged_df = pd.concat(
    stock_data.values(),
    axis=1
)

merged_df.head()
```

```python
merged_df.to_csv("../data/merged_stock_data.csv")
print("저장 완료!")
```
