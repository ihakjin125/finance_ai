```python
import requests
import os
import pandas as pd
from dotenv import load_dotenv
```

```python
# .env를 메모리에 올리기
load_dotenv("../.env")

# API 키 가져오기
client_id = os.getenv("NAVER_CLIENT_ID")
client_secret = os.getenv("NAVER_CLIENT_SECRET")

# URL 만들기
url = "https://openapi.naver.com/v1/search/news.json"
```

```python
# Header 만들기
headers = {
    "X-Naver-Client-Id": client_id,
    "X-Naver-Client-Secret": client_secret
}
```

```python
# 검색 조건 만들기
params = {
    "query": "삼성전자",
    "display": 5,
    "sort": "date"
}
```

```python
# 요청 보내기
response = requests.get(
    url,
    headers=headers,
    params=params
)
```

```python
# 200뜨면 네이버 서버 통신 연결 완료
print(response.status_code) 
```

```python
data = response.json()
```

```python
import re
from html import unescape

for news in data["items"]:

    title = re.sub(r"<.*?>", "", news["title"])
    description = re.sub(r"<.*?>", "", news["description"])

    title = unescape(title)
    description = unescape(description)

    print("제목 :", title)
    print("링크 :", news["link"])
    print("요약 :", description)
    print("-" * 60)
```

```python
news_list = []

for news in data["items"]:

    title = re.sub(r"<.*?>", "", news["title"])
    description = re.sub(r"<.*?>", "", news["description"])

    title = unescape(title)
    description = unescape(description)

    news_list.append({
        "제목": title,
        "링크": news["link"],
        "요약": description
    })
```

```python
print(news_list)
```

```python
news_df = pd.DataFrame(news_list)

news_df.head()
```

```python
news_df.to_csv("../data/naver_news.csv", index=False)
```
